```python
from scrapy import Selector
import re
from bs4 import BeautifulSoup as bs
import os
import shutil
import pandas as pd

import datetime
```

```python
# all files' path.
def eachfile(filepath):
    pathdir = os.listdir(filepath)
    for s in pathdir:
        newdir = os.path.join(filepath, s)
        if os.path.isfile(newdir): 
            if os.path.splitext(newdir)[1] == ".html":  
                soundfile.append(newdir)
        elif os.path.isdir(newdir):  
            eachfile(newdir)  
    return soundfile

def path_list(parent):
    xxx = []
    fp = parent
    os.chdir(fp)
    f = eachfile(fp)
    for i in range(len(f)):
        xxx.append(f[i])
    return xxx
```

```python
def para_key(path,keywords):
    page = open(path, 'r',encoding='utf-8').read()
    text = []
    
    # text in span
    sel = Selector(text = page)
    divs = sel.xpath('//div')
    for div in divs:
        li = div.xpath('font/text()').getall()+div.xpath('span/text()').getall()
        ps = [item.strip() for item in li]
        ps = ''.join(ps)
        if ps and len(ps)>20:
            text.append(ps)
    # text in p
    soup = bs(page,'html.parser')
    content_text = soup.select('p')
    for i in content_text:
        p = i.text.strip()
        if len(p)>20:
            text.append(p)
        
    # paragraph intergration in two pages
    para = []
    for i in range(len(text)-1):
        if text[i+1][0].islower():
            item = text[i]+' '+text[i+1]
            para.append(item)
        elif text[i][0].islower():
            pass
        else:
            para.append(text[i])
    if len(text)>=2:        
        if text[len(text)-1][0].isupper():
            para.append(text[len(text)-1])
         
    # get paragraphs contains keywords
    all_para = ''
    for pa in para:
        if any([j in pa for j in keywords]):
            all_para = all_para + pa + '\n---------------------------\n'
        else:
            pass
#     unique = list(set(all_para))
#     str = ' '
#     text = str.join(unique)    
    
    return all_para
#             print(pa)
#             print('\n----------------\n')
```

```python
keywords = ['COVID','covid','Covid','coronavirus','epidemic','pandemic','Epidemic','Pandemic','Cornavirus']
```
